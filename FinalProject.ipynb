They would really appreciate if one could predict for them who is gonna get churned so they can proactively go to the customer to provide them better services and turn customers' decisions in the opposite direction



## Business Questions ##
- Do customers with more dependents churn more?
- Does gender affect churn rate?
- Does marital status or age affect the churn rate?
- Do platinum/gold card holders churn less than blue card holders?
- Is there a specific age group with a higher churn rate?
- Is there sign that the customer is going to cancel their account?

## Key Insight: Transaction Amount & Churn

The **$3,000** spending mark appears to be a critical threshold — customers who spend below this amount are significantly more likely to churn.

Notably, the **$6,000–$7,000** range shows a churn rate of ~68%, and the **$9,000–$11,000** range shows an unusually high churn concentration, reaching 100% at the upper end — suggesting these may be outlier segments worth investigating further.

## Transaction Count Trend & Churn

Churned customers are heavily concentrated at **Q4/Q1 ratios below 1.0**, indicating a declining transaction frequency over time. Customers whose transaction count increased from Q1 to Q4 (ratio > 1.0) are significantly more likely to remain active, suggesting that an upward engagement trend is a strong retention signal.

```python
fig = px.imshow(bank_df.select_dtypes(include='number').corr(),
                title='Correlation Heatmap',
                color_continuous_scale='Tealgrn',
                text_auto='.2f')
fig.update_layout(width=1000, height=800)
fig.show()
```

*Findings of this HeatMap:*

Positive Corr:
1. Correlation between Months_On_Book and Customer_Age are 0.79 meaning that the older they get, the more they use their credit card.
2. Full correlation between Credit Limit and Avg_open_To_buy which means they're identical, one of them could be dropped.
3. Correlation between Total Transactions and Total Transactions Amount are 0.81 which is Intuitive.
4. Total_Revolving_Bal and Avg_Utilization_Ratio (0.62), higher balance means higher utilization, expected.

Negative Corr:
1. Credit_Limit and Avg_Utilization_Ratio (-0.48), customers with higher credit limits tend to use a smaller proportion of it.
2. Avg_Open_To_Buy and Avg_Utilization_Ratio (-0.54), same logic, more available credit = lower utilization

Notable Corr:
1. Total_Relationship_Count and Total_Trans_Amt (-0.35), customers with more products actually spend less per transaction, interesting.
2. Total_Relationship_Count and Total_Trans_Ct (-0.24), same pattern.

```python
fig = px.scatter(bank_df, x='Credit_Limit', y='Avg_Utilization_Ratio',
                 color='Attrition_Flag', title='Credit Limit vs Avg Utilization Ratio',
                 color_discrete_sequence=px.colors.qualitative.D3)
fig.show()
```

The less the Credit limit, the more the utilization ratio, confirmed by the -0.48 correlation on the previous heatmap.
High limit customers rarely churn.

A noticable amount of churned customers had 0 balance as shown by the yellow horizontal line where we rarely see any teal dots, that's an indicator that the 0 balance is an indicator that the customer is likely to churn.

**Customers with low credit limits and near-zero utilization are the highest churn risk — the bank should target this segment proactively with engagement offers or credit limit increases.

```python
fig = px.scatter(bank_df, x='Total_Trans_Amt', y='Total_Trans_Ct',
                 color='Attrition_Flag',
                 title='Transaction Amount vs Transaction Count by Attrition Flag',
                 color_discrete_sequence=px.colors.qualitative.D3)
fig.show()
```

- People who make less than **60** transactions and spend less that **3k** are the most likely to churn.
